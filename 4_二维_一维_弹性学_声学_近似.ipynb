# 线弹性和声学介质的二维和一维近似


虽然三维各向同性线性弹性运动方程是一种非常通用的介质描述，但它们的数值解在小型台式计算机或笔记本电脑上的计算要求相当高。因此，我们在本课中推导了三维线弹性运动方程和声学运动方程的二维和一维近似

## 二维线弹性介质近似

我们再次从各向同性线弹性介质的三维运动方程出发

\begin{align}
\rho\frac{\partial^2 u_1}{\partial t^2} &= \frac{\partial \sigma_{11}}{\partial x_1} + \frac{\partial \sigma_{12}}{\partial x_2} + \frac{\partial \sigma_{13}}{\partial x_3} + f_1 \nonumber \\
\rho\frac{\partial^2 u_2}{\partial t^2} &= \frac{\partial \sigma_{21}}{\partial x_1} + \frac{\partial \sigma_{22}}{\partial x_2} + \frac{\partial \sigma_{23}}{\partial x_3} + f_2\nonumber \\
\rho\frac{\partial^2 u_3}{\partial t^2} &= \frac{\partial \sigma_{31}}{\partial x_1} + \frac{\partial \sigma_{32}}{\partial x_2} + \frac{\partial \sigma_{33}}{\partial x_3} + f_3\nonumber \\
\sigma_{11} &= \lambda(\epsilon_{11}+\epsilon_{22}+\epsilon_{33}) + 2 \mu \epsilon_{11}\nonumber \\
\sigma_{22} &= \lambda(\epsilon_{11}+\epsilon_{22}+\epsilon_{33}) + 2 \mu \epsilon_{22}\nonumber \\
\sigma_{33} &= \lambda(\epsilon_{11}+\epsilon_{22}+\epsilon_{33}) + 2 \mu \epsilon_{33}\nonumber \\
\sigma_{12} &= 2 \mu \epsilon_{12}\nonumber \\
\sigma_{13} &= 2 \mu \epsilon_{13}\nonumber \\
\sigma_{23} &= 2 \mu \epsilon_{23}\nonumber \\
\end{align}

这一次我们假设材料参数不向某个方向变化。这种近似称为二维介质。二维介质的例子是运河、海堤或墙壁.

D对于介质的各向同性，我们可以选择任何方向，其中材料参数不发生变化。对于下面的推导，我们取两个方向。 在弹性的情况下, 我们可以从两个不同的方面来认识二维介质。首先, 我们可以设置所有的位移在两个方向 $u_2=0$ 同时这两个方向上的所有导数 $\frac{\partial}{\partial x_2}=0$. 这引出了 ...

### 二维 PSV 问题

为了推导二维PSV问题的偏微分方程, 我首先用红色标记所有由于o $\frac{\partial}{\partial x_2} = 0$ 而明显消失的项:

\begin{align}
\rho\frac{\partial^2 u_1}{\partial t^2} &= \frac{\partial \sigma_{11}}{\partial x_1} + \color{red}{\frac{\partial \sigma_{12}}{\partial x_2}} + \frac{\partial \sigma_{13}}{\partial x_3} + f_1 \nonumber \\
\rho\frac{\partial^2 u_2}{\partial t^2} &= \frac{\partial \sigma_{21}}{\partial x_1} + \color{red}{\frac{\partial \sigma_{22}}{\partial x_2}} + \frac{\partial \sigma_{23}}{\partial x_3} + f_2\nonumber \\
\rho\frac{\partial^2 u_3}{\partial t^2} &= \frac{\partial \sigma_{31}}{\partial x_1} + \color{red}{\frac{\partial \sigma_{32}}{\partial x_2}} + \frac{\partial \sigma_{33}}{\partial x_3} + f_3\nonumber \\
\sigma_{11} &= \lambda(\epsilon_{11}+\color{red}{\epsilon_{22}}+\epsilon_{33}) + 2 \mu \epsilon_{11}\nonumber \\
\sigma_{22} &= \lambda(\epsilon_{11}+\color{red}{\epsilon_{22}}+\epsilon_{33}) + \color{red}{2 \mu \epsilon_{22}}\nonumber \\
\sigma_{33} &= \lambda(\epsilon_{11}+\color{red}{\epsilon_{22}}+\epsilon_{33}) + 2 \mu \epsilon_{33}\nonumber \\
\sigma_{12} &= 2 \mu \epsilon_{12}\nonumber \\
\sigma_{13} &= 2 \mu \epsilon_{13}\nonumber \\
\sigma_{23} &= 2 \mu \epsilon_{23}\nonumber \\
\end{align}

注意, $\sigma_{22}$ 不再出现在动量方程中, 所以我们也可以忽略 $\sigma_{22}$ 方程，得到:


\begin{align}
\rho\frac{\partial^2 u_1}{\partial t^2} &= \frac{\partial \sigma_{11}}{\partial x_1} + \frac{\partial \sigma_{13}}{\partial x_3} + f_1\nonumber \\
\rho\frac{\partial^2 u_2}{\partial t^2} &= \frac{\partial \sigma_{21}}{\partial x_1} + \frac{\partial \sigma_{23}}{\partial x_3} + f_2\nonumber \\
\rho\frac{\partial^2 u_3}{\partial t^2} &= \frac{\partial \sigma_{31}}{\partial x_1} + \frac{\partial \sigma_{33}}{\partial x_3} + f_3\nonumber \\
\sigma_{11} &= \lambda(\epsilon_{11}+\epsilon_{33}) + 2 \mu \epsilon_{11}\nonumber \\
\sigma_{33} &= \lambda(\epsilon_{11}+\epsilon_{33}) + 2 \mu \epsilon_{33}\nonumber \\
\sigma_{12} &= 2 \mu \epsilon_{12}\nonumber \\
\sigma_{13} &= 2 \mu \epsilon_{13}\nonumber \\
\sigma_{23} &= 2 \mu \epsilon_{23}\nonumber \\
\end{align}

接下来，我们通过插入应变张量分量来计算剪应力:

\begin{align}
\color{red}{\sigma_{12}} &\color{red}{= \mu \biggl(\frac{\partial u_1}{\partial x_2} + \frac{\partial u_2}{\partial x_1}\biggr)}\nonumber \\
\sigma_{13} &= \mu \biggl(\frac{\partial u_1}{\partial x_3} + \frac{\partial u_3}{\partial x_1}\biggr)\nonumber \\
\color{red}{\sigma_{23}} &\color{red}{= \mu \biggl(\frac{\partial u_2}{\partial x_3} + \frac{\partial u_3}{\partial x_2}\biggr)}\nonumber \\
\end{align}

你注意到为什么 $\sigma_{12}$ and $\sigma_{23}$ (标红色) 为0? 

因为, $u_2 = 0$ 和 $\frac{\partial}{\partial x_2}=0$. 因此, 所有 $\sigma_{12}$ and $\sigma_{23}$ 项为0, 正如我们第一次标红色的:

\begin{align}
\rho\frac{\partial^2 u_1}{\partial t^2} &= \frac{\partial \sigma_{11}}{\partial x_1} + \frac{\partial \sigma_{13}}{\partial x_3} + f_1\nonumber \\
\color{red}{\rho\frac{\partial^2 u_2}{\partial t^2}} &\color{red}{= \frac{\partial \sigma_{21}}{\partial x_1} + \frac{\partial \sigma_{23}}{\partial x_3}} + f_2 \nonumber \\
\rho\frac{\partial^2 u_3}{\partial t^2} &= \frac{\partial \sigma_{31}}{\partial x_1} + \frac{\partial \sigma_{33}}{\partial x_3} + f_3\nonumber \\
\sigma_{11} &= \lambda(\epsilon_{11}+\epsilon_{33}) + 2 \mu \epsilon_{11}\nonumber \\
\sigma_{33} &= \lambda(\epsilon_{11}+\epsilon_{33}) + 2 \mu \epsilon_{33}\nonumber \\
\color{red}{\sigma_{12}} &\color{red}{= 2 \mu \epsilon_{12}}\nonumber \\
\sigma_{13} &= 2 \mu \epsilon_{13}\nonumber \\
\color{red}{\sigma_{23}} &\color{red}{= 2 \mu \epsilon_{23}}\nonumber \\
\end{align}

删除掉红色的项:

\begin{align}
\rho\frac{\partial^2 u_1}{\partial t^2} &= \frac{\partial \sigma_{11}}{\partial x_1} + \frac{\partial \sigma_{13}}{\partial x_3} + f_1\nonumber \\
\rho\frac{\partial^2 u_3}{\partial t^2} &= \frac{\partial \sigma_{31}}{\partial x_1} + \frac{\partial \sigma_{33}}{\partial x_3} + f_3\nonumber \\
\sigma_{11} &= \lambda(\epsilon_{11}+\epsilon_{33}) + 2 \mu \epsilon_{11}\nonumber \\
\sigma_{33} &= \lambda(\epsilon_{11}+\epsilon_{33}) + 2 \mu \epsilon_{33}\nonumber \\
\sigma_{13} &= 2 \mu \epsilon_{13}\nonumber \\
\end{align}

最后，我们可以重新排列应变张量分量:

\begin{align}
\rho\frac{\partial^2 u_1}{\partial t^2} &= \frac{\partial \sigma_{11}}{\partial x_1} + \frac{\partial \sigma_{13}}{\partial x_3} + f_1\nonumber \\
\rho\frac{\partial^2 u_3}{\partial t^2} &= \frac{\partial \sigma_{31}}{\partial x_1} + \frac{\partial \sigma_{33}}{\partial x_3} + f_3\nonumber \\
\sigma_{11} &= (\lambda+2\mu)\epsilon_{11} + \lambda \epsilon_{33}\nonumber \\
\sigma_{33} &= \lambda \epsilon_{11} + (\lambda + 2 \mu) \epsilon_{33}\nonumber \\
\sigma_{13} &= 2 \mu \epsilon_{13}\nonumber \\
\end{align}

通过插入应变张量分量，得到 **应力-位移公式** 中 **二维PSV问题** 的偏微分方程

\begin{align}
\rho\frac{\partial^2 u_1}{\partial t^2} &= \frac{\partial \sigma_{11}}{\partial x_1} + \frac{\partial \sigma_{13}}{\partial x_3} + f_1\nonumber \\
\rho\frac{\partial^2 u_3}{\partial t^2} &= \frac{\partial \sigma_{31}}{\partial x_1} + \frac{\partial \sigma_{33}}{\partial x_3} + f_3\nonumber \\
\sigma_{11} &= (\lambda+2\mu)\frac{\partial u_1}{\partial x_1} + \lambda \frac{\partial u_3}{\partial x_3}\nonumber \\
\sigma_{33} &= \lambda \frac{\partial u_1}{\partial x_1} + (\lambda + 2 \mu) \frac{\partial u_3}{\partial x_3}\nonumber \\
\sigma_{13} &= \mu \biggl(\frac{\partial u_1}{\partial x_3} + \frac{\partial u_3}{\partial x_1}\biggr)\nonumber \\
\end{align}

利用二维PSV方程，我们可以模拟P波和SV波的传播，并假设正确的边界条件，瑞利波.

除了假设 $u_2$ 分量为零, 我们还可以设置 $u_1=0$, $u_3=0$ 和 $\frac{\partial}{\partial x_2} = 0$, 而 $u_2$ 是唯一的非零分量. 这引出了 ...


### 二维SH问题

像PSV问题, 我们首先假定 $\frac{\partial}{\partial x_2} = 0$. 这会导致 

\begin{align}
\rho\frac{\partial^2 u_1}{\partial t^2} &= \frac{\partial \sigma_{11}}{\partial x_1} + \frac{\partial \sigma_{13}}{\partial x_3} + f_1\nonumber \\
\rho\frac{\partial^2 u_2}{\partial t^2} &= \frac{\partial \sigma_{21}}{\partial x_1} + \frac{\partial \sigma_{23}}{\partial x_3} + f_2\nonumber \\
\rho\frac{\partial^2 u_3}{\partial t^2} &= \frac{\partial \sigma_{31}}{\partial x_1} + \frac{\partial \sigma_{33}}{\partial x_3} + f_3\nonumber \\
\sigma_{11} &= \lambda(\epsilon_{11}+\epsilon_{33}) + 2 \mu \epsilon_{11}\nonumber \\
\sigma_{33} &= \lambda(\epsilon_{11}+\epsilon_{33}) + 2 \mu \epsilon_{33}\nonumber \\
\sigma_{12} &= 2 \mu \epsilon_{12}\nonumber \\
\sigma_{13} &= 2 \mu \epsilon_{13}\nonumber \\
\sigma_{23} &= 2 \mu \epsilon_{23}\nonumber \\
\end{align}

因为 $u_1=0$, $u_3=0$ 和 $\frac{\partial}{\partial x_2} = 0$，我们可以删除以下应力项: 

\begin{align}
\color{red}{\sigma_{11}} &\color{red}{= \lambda(\epsilon_{11}+\epsilon_{33}) + 2 \mu \epsilon_{11}}\nonumber \\
\color{red}{\sigma_{33}} &\color{red}{= \lambda(\epsilon_{11}+\epsilon_{33}) + 2 \mu \epsilon_{33}}\nonumber \\
\sigma_{12} &= 2 \mu \epsilon_{12}\nonumber \\
\color{red}{\sigma_{13}} &\color{red}{= 2 \mu \epsilon_{13}}\nonumber \\
\sigma_{23} &= 2 \mu \epsilon_{23}\nonumber \\
\end{align}

因此, 这些动量方程也变为了0:

\begin{align}
\color{red}{\rho\frac{\partial^2 u_1}{\partial t^2}} &\color{red}{= \frac{\partial \sigma_{11}}{\partial x_1} + \frac{\partial \sigma_{13}}{\partial x_3} + f_1}\nonumber \\
\rho\frac{\partial^2 u_2}{\partial t^2} &= \frac{\partial \sigma_{21}}{\partial x_1} + \frac{\partial \sigma_{23}}{\partial x_3} + f_2\nonumber \\
\color{red}{\rho\frac{\partial^2 u_3}{\partial t^2}} &= \color{red}{\frac{\partial \sigma_{31}}{\partial x_1} + \frac{\partial \sigma_{33}}{\partial x_3} + f_3}\nonumber
\end{align}

剩下的方程为:

\begin{align}
\rho\frac{\partial^2 u_2}{\partial t^2} &= \frac{\partial \sigma_{21}}{\partial x_1} + \frac{\partial \sigma_{23}}{\partial x_3} + f_2\nonumber \\
\sigma_{12} &= 2 \mu \epsilon_{12}\nonumber \\
\sigma_{23} &= 2 \mu \epsilon_{23}\nonumber \\
\end{align}

插入 $\epsilon_{12}$ 和 $\epsilon_{23}$ 后, 我们得到: 

\begin{align}
\rho\frac{\partial^2 u_2}{\partial t^2} &= \frac{\partial \sigma_{21}}{\partial x_1} + \frac{\partial \sigma_{23}}{\partial x_3} + f_2\nonumber \\
\sigma_{12} &= \mu \biggl(\color{red}{\frac{\partial u_1}{\partial x_2}} + \frac{\partial u_2}{\partial x_1}\biggr)\nonumber \\
\sigma_{23} &= \mu \biggl(\frac{\partial u_2}{\partial x_3} + \color{red}{\frac{\partial u_3}{\partial x_2}}\biggr)\nonumber \\
\end{align}

这里, $\frac{\partial u_1}{\partial x_2}$ 和 $\frac{\partial u_3}{\partial x_2}$ 均为0, 我们最终得到了 **二维SH问题** 的偏微分方程在 **位移-应力公式**

\begin{align}
\rho\frac{\partial^2 u_2}{\partial t^2} &= \frac{\partial \sigma_{21}}{\partial x_1} + \frac{\partial \sigma_{23}}{\partial x_3} + f_2\nonumber \\
\sigma_{12} &= \mu \frac{\partial u_2}{\partial x_1}\nonumber \\
\sigma_{23} &= \mu \frac{\partial u_2}{\partial x_3}\nonumber \\
\end{align}

利用这些方程，我们可以描述水平剪切波(SH)和Love波的传播.

##### 注意！

由于介质是双向无限扩展的，与现场数据应用中使用的爆炸源或冲击源相比，上述二维弹性介质近似中激发的所有源 $f_i$ 这些线源不辐射球形波, 除了**圆柱波**. 

因此，模拟的辐射源模式和几何扩散效应与现场实测的不同。这些差异必须得到补偿，例如，如果我们想通过二维全波形反演从现场数据集重建地下属性。

## 一维SH弹性介质近似

二维SH问题已经很紧凑了:

\begin{align}
\rho\frac{\partial^2 u_2}{\partial t^2} &= \frac{\partial \sigma_{21}}{\partial x_1} + \frac{\partial \sigma_{23}}{\partial x_3} + f_2\nonumber \\
\sigma_{12} &= \mu \frac{\partial u_2}{\partial x_1}\nonumber \\
\sigma_{23} &= \mu \frac{\partial u_2}{\partial x_3}\nonumber \\
\end{align}

因此在计算上不像三维各向同性弹性运动方程那样费力。然而，对于地震模型的介绍我们来看看，能否摆脱另一个维度。在把应力张量分量插入动量方程之后, 我们得到

\begin{align}
\rho\frac{\partial^2 u_2}{\partial t^2} &= \frac{\partial}{\partial x_1} \mu \frac{\partial u_2}{\partial x_1} + \frac{\partial}{\partial x_3} \mu \frac{\partial u_2}{\partial x_3} + f_2\nonumber \\
\end{align}

假设剪切模量 $\mu$ 和 $u_2$ 在3个方向上没有变化, 或者 $\frac{\partial}{\partial x_3} = 0$, 我们得到 **1D SH problem**

\begin{align}
\rho\frac{\partial^2 u_2}{\partial t^2} &= \frac{\partial}{\partial x_1} \mu \frac{\partial u_2}{\partial x_1} + f_2\nonumber \\
\end{align}

二维介质中的点源辐射柱面波, 一维介质近似中的源辐射 **平面波**.


## 二维和一维声学介质近似

二维和一维声学介质近似很容易引入, 因为我们已经推到了 **3D acoustic wave equation** in [Lesson 3: Acoustic medium approximation](http://nbviewer.jupyter.org/github/daniel-koehn/Theory-of-seismic-waves-II/blob/master/01_Analytical_solutions/3_Acoustic_medium.ipynb):

\begin{equation}
\frac{1}{V_p^2} \frac{\partial^2 P}{\partial t^2} = \frac{\partial^2 P}{\partial x_1^2} + \frac{\partial^2 P}{\partial x_2^2} + \frac{\partial^2 P}{\partial x_3^2}\nonumber
\end{equation}

通过设置 $\frac{\partial}{\partial x_2} = 0$, 我们得到 **2D acoustic wave equation** 

\begin{equation}
\frac{1}{V_p^2} \frac{\partial^2 P}{\partial t^2} = \frac{\partial^2 P}{\partial x_1^2} + \frac{\partial^2 P}{\partial x_3^2}\nonumber
\end{equation}

通过设置 $\frac{\partial}{\partial x_3} = 0$, 我们得到 **1D acoustic wave equation**

\begin{equation}
\frac{1}{V_p^2} \frac{\partial^2 P}{\partial t^2} = \frac{\partial^2 P}{\partial x_1^2}\nonumber
\end{equation}
